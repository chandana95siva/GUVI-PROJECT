```python
path = "/content/pulse/data/aggregated/transaction/country/india/state/"
Agg_state_list = os.listdir(path)
Agg_state_list

clm = {'State':[],'Year':[],'Quater':[],'Transaction_type':[],'Transaction_count':[],'Transaction_amt':[]}
for i in Agg_state_list:
  P_i = path+i+"/"
  Agg_yr = os.listdir(P_i)
  for j in Agg_yr:
    P_j =  P_i+j+"/"
    Agg_yrlist =os.listdir(P_j)
    for k in Agg_yrlist:
      P_k = P_j+k
      data = open(P_k,'r')
      D = json.load(data)
      for l in D['data']['transactionData']:
        Name = l['name']
        Amount = l['paymentInstruments'][0]['amount']
        Count = l['paymentInstruments'][0]['count']
        clm['Transaction_type'].append(Name)
        clm['Transaction_amt'].append(Amount)
        clm['Transaction_count'].append(Count)
        clm['State'].append(i)
        clm['Year'].append(j)
        clm['Quater'].append(int(k.strip('.json')))
AGG_TRANS = pd.DataFrame(clm)
```

```python
path = "/content/pulse/data/aggregated/user/country/india/state/"
Agg_state_list = os.listdir(path)
Agg_state_list
Clm = {'State':[],'Year':[],'Quater':[],'Brand_name': [],'UserCount':[],'UserPercentage':[]}
for i in Agg_state_list:
  p_i = path+i+"/"
  Agg_yr = os.listdir(p_i)
  for j in Agg_yr:
    P_j =  p_i+j+"/"
    Agg_yrlist =os.listdir(P_j)
    for k in Agg_yrlist:
      P_k = P_j+k
      data = open(P_k,'r')
      D = json.load(data)
      try:
        for u in D['data']['usersByDevice']:
          Brand = u['brand']
          Count = u['count']
          Percentage = u['percentage']
          Clm['Brand_name'].append(Brand)
          Clm['UserCount'].append(Count)
          Clm['UserPercentage'].append(Percentage)
          Clm['State'].append(i)
          Clm['Year'].append(j)
          Clm['Quater'].append(int(k.strip('.json')))
      except:
        pass
AGG_User = pd.DataFrame(Clm)
```

```python
path = "/content/pulse/data/map/transaction/hover/country/india/state/"
Agg_state_list = os.listdir(path)
Agg_state_list

clm = {'State':[],'Year':[],'Quater':[],'metric_amount':[],'metric_count':[],'metric_type':[]}
for i in Agg_state_list:
  P_i = path+i+"/"
  Agg_yr = os.listdir(P_i)
  for j in Agg_yr:
    P_j =  P_i+j+"/"
    Agg_yrlist =os.listdir(P_j)
    for k in Agg_yrlist:
      P_k = P_j+k
      data = open(P_k,'r')
      D = json.load(data)
      for H in D['data']['hoverDataList']:
        Name = H['name']
        Amount = H['metric'][0]['amount']
        Count = H['metric'][0]['count']
        clm['metric_type'].append(Name)
        clm['metric_amount'].append(Amount)
        clm['metric_count'].append(Count)
        clm['State'].append(i)
        clm['Year'].append(j)
        clm['Quater'].append(int(k.strip('.json')))
MAP_TRANS = pd.DataFrame(clm)
```

```python
path = "/content/pulse/data/map/user/hover/country/india/state/"
Agg_state_list = os.listdir(path)
Agg_state_list

clm = {'State':[],'Year':[],'Quater':[],'RegisteredUsers':[],'DISTRICTS':[]}
for i in Agg_state_list:
  P_i = path+i+"/"
  Agg_yr = os.listdir(P_i)
  for j in Agg_yr:
    P_j =  P_i+j+"/"
    Agg_yrlist =os.listdir(P_j)
    for k in Agg_yrlist:
      P_k = P_j+k
      data = open(P_k,'r')
      D = json.load(data)
      for H in D['data']['hoverData'].items():
        Districts = H[0]
        Registereduser = H[1]['registeredUsers']
        clm['DISTRICTS'].append(Districts)

        clm['RegisteredUsers'].append(Registereduser)
        clm['State'].append(i)
        clm['Year'].append(j)
        clm['Quater'].append(int(k.strip('.json')))
MAP_USER = pd.DataFrame(clm)
```
